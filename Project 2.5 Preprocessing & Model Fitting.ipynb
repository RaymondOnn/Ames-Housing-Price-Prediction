# Part 5: Data Preprocessing and Model training

## Importing Libraries

```python
# import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
import category_encoders as ce

from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, Ridge
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.metrics import r2_score, mean_squared_error

%matplotlib inline
%config InlineBackend.figure_format = 'retina' # Render high resolution images

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.float_format', lambda x: '%.5f' % x)  # switch off scientic notation
```

## Loading data back

```python
# Load data
raw = pd.read_csv('data/transformed_data part 3.csv',
                  na_values="",  # There are no missing values in the clean data file.
                  keep_default_na=False  # "NA" strings are casted as actual values.
                 )
raw.head()
```

```python
raw.info()
```

```python
# loading lists
with open('data/columns v0.4.txt', 'r', encoding='utf-8') as f:
    col_dict = json.load(f)

# Use one or the other as dependent variable
target = ['SalePrice']
log_target = ['log-SalePrice']

# Use cont or log_cont. Do not use together -> collinearity issues
cont = ['Lot Area', '1st Flr SF', 'Gr Liv Area'] 
log_cont = ['log-Lot Area', 'log-1st Flr SF', 'log-Gr Liv Area'] 

# Do not use in model fitting
del_cols = ['Id', 'PID', 'Year Built', 'Year Remod/Add', '2nd Flr SF', 'BsmtFin SF 1' ]  

cols = list(set(raw.columns.to_list()) - set(target) - set(log_target) - set(cont) - set(log_cont) - set(del_cols))
data = raw.copy()
```

```python
# Unique values for each column
for item in sorted(data.columns.to_list()):
    print (item, ": ", data[item].nunique(),"values")
    print (data[item].unique(), "\n")
```

```python
# Prepping features and response variables
x = data[cols + cont]
y = np.ravel(data[log_target])
features = x.columns
assert not 'log-SalePrice' in features.to_list()
print(x.shape)
```

## Variance Thresholding

```python
# Filter out low variance columns
variance_threshold = 0.01 # set threshold here
vt = VarianceThreshold(threshold=variance_threshold) # create instance of variance threshold
vt.fit(x)
x_selected = pd.DataFrame(vt.transform(x), columns=x.columns[vt.get_support()]) # load selected columns into dataframe
features_selected = x_selected.columns
x_selected.head()
x_selected.shape
```

## Preprocessing

### Train test split

```python
# Splitting the data into training and testing portions
x_train, x_test, y_train, y_test = train_test_split(x_selected, y, test_size = 0.30, random_state = 1)
```

### Standard Scaling

```python
# Scaling the continuous variables
ss = StandardScaler()
scaler = ss.fit(x_train)    # Compute the mean and std to be used for later scaling.
x_train_ss = scaler.transform(x_train) # Perform standardization by centering and scaling
x_test_ss = scaler.transform(x_test)
```

```python
# Create an instance of the models
lr = LinearRegression()
# ridge = Ridge(alpha=10) # RidgeCV(alphas=np.linspace(.1, 10, 100))
# lasso = LassoCV(n_alphas=200)
```

## Model Fitting

```python
def train_model(model, name):
    
    model.fit(x_train_ss, y_train)
    r2_scores = cross_val_score(model, x_train_ss, y_train, scoring='r2',cv=5)
    mse_scores = cross_val_score(model, x_train_ss, y_train, scoring='neg_mean_squared_error',cv=5)

    # Step 3. Generate predictions.
    preds = model.predict(x_test_ss)


    print("\n\nRMSE & R2 SCORE: {}".format(name))
    print("-"*100)
    
    # Check the MSE on the training and testing sets.
    #print(f'RMSE on training set: {np.sqrt(mean_squared_error(y_train, model.predict(x_train_ss)))}')
    print(f'RMSE on training set: {np.sqrt(mse_scores.mean() * -1)}')
    print(f'RMSE on testing set: {np.sqrt(mean_squared_error(y_test, preds))}','\n')

    # Check the R^2 on the training and testing sets.
    print(f'R^2 on training set: {r2_scores.mean()}')
    print(f'R^2 on testing set: {r2_score(y_test, preds)}')         
    print("-"*100)
    
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16, 7))
    residuals = y_test - preds
    ax1.scatter(preds, residuals)
    ax1.set_title('Homoscedasticity Check', fontsize=14)
    ax1.set_xlabel('Predictions')
    ax1.set_ylabel('Residuals')
    ax1.axhline(0, color='orange')
    ax2.hist(residuals, bins=100)
    ax2.set_title('Residual Normality Check', fontsize=14)
```

### Fitting the OLS Model

```python
train_model(lr, 'LinReg')

# display(pd.Series(np.ravel(lr.coef_), index=features_selected).to_frame('Coef').sort_values('Coef'))
```

### Fitting the Ridge Model

#### Optimizing alpha for Ridge model

```python
# Set up a list of ridge alphas to check.
# np.logspace generates 100 values equally between 0 and 5,
# then converts them to alphas between 10^0 and 10^5.
r_alphas = np.logspace(0, 5, 100)

# Cross-validate over our list of ridge alphas.
ridge_cv = RidgeCV(alphas=r_alphas, scoring='r2', cv=5)

# Fit model using best ridge alpha!
ridge_best = ridge_cv.fit(x_train_ss, y_train)

# Here is the optimal value of alpha
a = ridge_best.alpha_
print(a)

r_alphas = np.linspace(a-50, a+50, 100)
ridge_cv = RidgeCV(alphas=r_alphas, scoring='r2', cv=5)
ridge_best = ridge_cv.fit(x_train_ss, y_train)
a = ridge_best.alpha_
print(a)
```

#### Model Fitting

```python
train_model(ridge_best, 'Ridge')

#print(np.ravel(ridge_best.coef_).shape)
#res_df = pd.Series(np.ravel(ridge_best.coef_), index=features_selected).sort_values().to_frame('coef')
#res_df[res_df['coef'] != 0]
```

### Fitting the Lasso Model

#### Optimizing alpha for Lasso model

```python
# Set up a list of Lasso alphas to check.
l_alphas = np.logspace(-3, 0, 100)

# Cross-validate over our list of Lasso alphas.
lasso_cv = LassoCV(alphas=l_alphas, cv=5, max_iter=50000)

# Fit model using best ridge alpha!
lasso_best = lasso_cv.fit(x_train_ss, y_train)

# Here is the optimal value of alpha
a = lasso_best.alpha_
print(a)

l_alphas = np.linspace(a-0.5, a+0.5, 100)
lasso_cv = LassoCV(alphas=l_alphas, cv=5, max_iter=50000)
lasso_best = lasso_cv.fit(x_train_ss, y_train)
a = lasso_best.alpha_
print(a)
```

#### Model Fitting

```python
train_model(lasso_best, 'Lasso' )
#res_df = pd.Series(np.ravel(lasso_best.coef_), index=features_selected).sort_values().to_frame('coefficient')
#res_df[res_df['coefficient'] != 0]
```

## Results & Analysis

A total of 3 regression models were evaluated. The models considered were the normal linear regression, Ridge regression and the LASSO regression. Evaluation was done on the basis of R-squared and Root Mean Squared Error(RMSE)

Prior to model fitting, the dataset was first put through a variance threshold with the threshold set at 0.01, allowing features with variance 0.01 or higher to pass through. The features are normalized using the Standard Scaler after train test split.

During, model fitting, 5-fold cross validation was done on the train dataset for all three models. Subsequently, the models were then tested using the test dataset

All in all, The Ridge model was selected as the model of choice as it scored the highest R-Squared and the lowest Root Mean Squared Error(RMSE) on the testing dataset. Test R-squared score was 0.905 and the test RMSE was 0.1246.
The Ridge model also outperformed the Null model, which gave a baseline R-Squared of 0.00 and a baseline RMSE of 0.4111

The 5 biggest determinants of housing sale price were found to be 
* Overall Quality 
* Above Ground Living Area, 
* having at least average condition, 
* Total rooms above ground
* the 1st Floor Area

This makes sense as buyers do not want to rebuild the house all over again. Hence, they generally expect the house of be of a certain quality and condition. 
Above ground living area, total rooms above ground and 1st floor area can be seen as proxy for the land area of the property as by buying the property, buyers essentially have ownership over the land for which the property sits.
The more land you purchase, the higher the price.

# Part 6: Generate Predictions for Submissions

## Load Test Data

```python
# import data
df = pd.read_csv('data/test.csv', 
                  na_values="",  # There are no missing values in the clean data file.
                  keep_default_na=False  # "NA" strings are casted as actual values.
                 )
```

## Feature Engineering

### Age of House / Age of Remodelling

```python
df['yr_since_built'] = df['Yr Sold'] - df['Year Built']
df['yr_since_remod'] = df['Yr Sold'] - df['Year Remod/Add']

# Houses that are built in the recent 10 years
df['is_built_recently'] = np.where(df['yr_since_built'] < 11, 1, 0)

# Houses that are remodeled in the recent 10 years
df['is_remod_recently'] = np.where(df['yr_since_remod'] < 11, 1, 0)
```

### Alley

```python
df['has_alley'] = np.where(df['Alley']=='NA', 0, 1)
df['is_pave_alley'] = np.where(df['Alley']=='Pave', 1, 0)
```

### Splitting Condition 1 /  Condition 2

#### Major street
* 0: No, 
* 1: Yes

```python
# Based on the values, Condition 1 looks like a combination of features i.e. distance from different locations (Arterial street, feeder street, North-South Railroad, East-West Railroad, off-site feature
# Hence, this column should be split into the various columns with {'Normal': 0, 'Within 200m': 1, 'Adjacent': 2}}

maj_str = ['Artery', 'Feedr']
# Arterial street: (0: Normal, 1: Artery-Adjacent to arterial street
df['near_major_street'] = df.apply(lambda x: 1 if (x['Condition 1'] in maj_str) | (x['Condition 2'] in maj_str) else 0, axis='columns')
#df['near_art_st'] = df.apply(lambda x: 'NO' if (x['Condition 1'] != 1.0) & (x.participation_sat != 1.0) else ( "ACT" if (x.participation_act == 1.0)  else "SAT"), axis='columns')
df['near_major_street'].unique()
```

#### Railroad
* 0: No, 
* 1: Yes, Within 200' or Adjacent to Railroad

```python
# Would be preferable to distinct between 'within 200m' and 'adjacent'
# However, given the low number of sample, probably better to add them up for more weight 
rail = ['RRAn', 'RRNn', 'RRAe', 'RRNe']
df['near_rail'] = df.apply(lambda x: 1 if (x['Condition 1'] in rail) | (x['Condition 2'] in rail) else 0, axis='columns')
df['near_rail'].unique()
df[['near_rail', 'Condition 1', 'Condition 2']].head(20)
```

#### Off-Site feature 
* 0: Normal, 
* 1: PosN Near positive off-site feature--park, greenbelt, etc., 
* 2: PosA Adjacent to postive off-site feature

```python
# Create new column
df['near_offsite'] = df.apply(lambda x: 
                              2 if (x['Condition 1'] == 'PosA') | (x['Condition 2'] == 'PosA') else ( 
                                  1 if (x['Condition 1'] == 'PosN') | (x['Condition 2'] == 'PosN')  else 
                                  0), axis='columns')
df['near_offsite'].unique()
cols_to_be_dropped = ['Condition 1', 'Condition 2']
```

### Splitting Exterior 1st /  Exterior 2nd

```python
lst = ['VinylSd',	'MetalSd',	'HdBoard',	'Wd Sdng',	'Plywood',	'CemntBd',	'BrkFace',	'WdShing',	'AsbShng',	'Stucco',	'BrkComm',	'Stone',	'CBlock',	'ImStucc',	'AsphShn']
for val in lst:
    new_col = 'ext_' + val
    df[new_col] = df.apply(lambda x: 1 if (x['Exterior 1st'] == val)|( x['Exterior 2nd'] == val) else 0, axis='columns')
```

### Central Air

```python
# Create new column
df['has_aircon'] = np.where(df['Central Air'] == 'Yes', 1, 0)
```

### Electrical

```python
df['has_std_cb'] = np.where(df['Electrical'] == 'SBrkr', 1, 0)
```

### Garage Type & Finish

```python
lst = ['Attchd', 'BuiltIn']
df['is_garage_attached'] = np.where(df['Garage Type'].isin(lst), 1, 0)
df[['is_garage_attached', 'Garage Type']].head(20)
```

```python
df['has_garage'] = np.where(df['Garage Finish'] == 'NA', 0, 1)
df['is_garage_finished'] = np.where(df['Garage Finish'].isin(['Fin', 'RFn']), 1, 0)
```

### Lot Shape

```python
df['is_reg_lot'] = np.where(df['Lot Shape'] == 'Reg', 1, 0)
```

### Overall Cond

```python
df['at_least_avg_cond'] = np.where(df['Overall Cond'] < 5, 0, 1)
```
